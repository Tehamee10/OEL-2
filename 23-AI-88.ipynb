Implement BFS

```python
import collections

def bfs(graph,root,goalstate):
  visited = set()
  queue = collections.deque([root])
  sum = 0
  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      sum+=vertex[1]
      visited.add(vertex)
    if vertex[0] == goalstate:
      print("The Total distance between A to G is:",sum)
      return visited


    for edge in graph[vertex[0]]:
      if edge not in visited:
         queue.append(edge)
  return visited

graph = {
 'A': [('B', 6), ('C', 2)],
 'B': [('D', 5), ('E', 3)],
 'C': [('F', 4)],
 'D': [('G', 2)],
 'E': [('G', 6)],
 'F': [('G', 1)],
 'G': []
}
goalstate = 'G'
root = ('A',0)
print(bfs(graph,root,goalstate))
```

```python
import collections

def dfs(graph,root,goalstate):
  visited = set()
  stack = collections.deque([root])
  sum = 0
  while stack:
    vertex = stack.pop()
    if vertex not in visited:
      sum+=vertex[1]
      visited.add(vertex)
    if vertex[0] == goalstate:
      print("The Total distance between A to G is:",sum)
      return visited


    for edge in graph[vertex[0]]:
      if edge not in visited:
         stack.append(edge)
  return visited

graph = {
 'A': [('B', 6), ('C', 2)],
 'B': [('D', 5), ('E', 3)],
 'C': [('F', 4)],
 'D': [('G', 2)],
 'E': [('G', 6)],
 'F': [('G', 1)],
 'G': []
}
goalstate = 'G'
root = ('A',0)
print(bfs(graph,root,goalstate))
```
